```python
df2 = df.iloc[2:228,1:51].copy().reset_index(drop=True)
```

```python
df2.head(15)
```

```python
df2.rename(columns={'Unnamed: 1':'Country'}, inplace=True)
```

```python
df2.head()
```

```python
df3 = df2[["Country","2021"]].copy()
```

```python
df3.head(15)
```

```python
df3.info()
```

```python
df3['2021'] = df3['2021'].replace(['--'], '0')
```

```python
df3['2021'] = df3['2021'].fillna(0)
```

```python
df3.rename(columns={'2021':'Production'}, inplace=True)
```

```python
df3.info()
```

```python
df3.shape
```

```python
df4= df3.copy()
```

```python
df4['Production'] = df4['Production'].astype(float)
```

```python
df4.info()
```

```python
df4.head()
```

```python
df4["Ranking"]= df4["Production"].rank(ascending=False)
```

```python
df4.head()
```

```python
df5= df4.sort_values("Ranking", ascending=True).reset_index(drop=True)
```

```python
df5.head(20)
```

```python
df5.tail()
```

```python
df5["Production"]=df5["Production"]/1000
```

```python
df5.head(11)
```

```python
df5["Production"]= df5["Production"].round(decimals =2)
```

```python
df5.head(12)
```

```python
df_top_ten_producers = df5.query('Ranking <= 10')
```

```python
df_top_ten_producers.head(11)
```

```python
df_top_ten_producers["Production"].describe()
```

```python
df_top_ten_producers.to_csv('Top_10_Oil_Producer_13_03_2022.csv',index=False)
```
